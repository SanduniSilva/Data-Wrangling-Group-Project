```python
import pandas as pd

# Import data set
fb = pd.read_excel('D:/Uni/5 Semester/Data Wrangling/final/Facebook-Google+andNewsStory.xlsx')
missing = pd.read_excel('D:/Uni/5 Semester/Data Wrangling/final/Missing Publication Date.xlsx')
pereview = pd.read_excel('D:/Uni/5 Semester/Data Wrangling/final/peerreview-patent and policy data.xlsx')
subcode = pd.read_excel('D:/Uni/5 Semester/Data Wrangling/final/Subject codes.xlsx')
tweet = pd.read_excel('D:/Uni/5 Semester/Data Wrangling/final/Tweet Data.xlsx')
```

```python
# Summary of the data set
print("Column names of FB data set \n", fb.columns.tolist())
print("Column names of Missing data set \n", missing.columns.tolist())
print("Column names of Preview data set \n", pereview.columns.tolist())
print("Column names of Subject Code data set \n", subcode.columns.tolist())
print("Column names of Tweet data set \n", tweet.columns.tolist())
```

```python
# Merge data sets into one data frame
merge_df = pd.concat([fb, missing, pereview, subcode, tweet])
print("Column names of Merge data set \n", merge_df.columns.tolist())

print(merge_df.shape[0])
```

```python
columns_to_clean = ['SubjectCode1', 'SubjectCode2', 'SubjectCode3']
merge_df.dropna(subset=columns_to_clean, inplace=True)

merge_df.to_csv('clean_null.csv', index=False)
print(merge_df.head)
print(merge_df.shape[0])
```

```python
# Function to replace SubjectCode with Subject names
def replace_subject_code(row):
    try:
        row['SubjectCode1'] = subcode.loc[subcode['Subject Code'] == row['SubjectCode1'], 'Subject'].iloc[0]
    except IndexError:
        row['SubjectCode1'] = None
        
    try:
        row['SubjectCode2'] = subcode.loc[subcode['Subject Code'] == row['SubjectCode2'], 'Subject'].iloc[0]
    except IndexError:
        row['SubjectCode2'] = None
        
    try:
        row['SubjectCode3'] = subcode.loc[subcode['Subject Code'] == row['SubjectCode3'], 'Subject'].iloc[0]
    except IndexError:
        row['SubjectCode3'] = None

    return row

# Perform the join and replace operation
rename_df = merge_df.apply(replace_subject_code, axis=1)
 
# Csv file 
rename_df.to_csv('rename_df.csv', index=False)
print("Column names of Rename data set \n", rename_df.columns.tolist())

print(rename_df.shape[0])
```

```python
# Define a custom function to merge the subject codes
def merge_subject_codes(row):
    codes = [str(row["SubjectCode1"]), str(row["SubjectCode2"]), str(row["SubjectCode3"])]
    merged_codes = "; ".join(code for code in codes if code != "")
    return merged_codes

# Apply the custom function to create the new merged column
rename_df["Subjects"] = rename_df.apply(merge_subject_codes,axis=1)

rename_df.to_csv('rename_merge.csv', index=False)
print("Column names of Rename data set \n", rename_df.columns.tolist())
print(rename_df.shape[0])
```

```python
# Create a new column of mention timme
rename_df['Mention Time'] = pd.to_datetime(rename_df['Mention Date']).dt.time
# Remove the time portion from the "Mention Date" column
rename_df['Mention Date'] = pd.to_datetime(rename_df['Mention Date']).dt.date

print(rename_df.shape[0])
rename_df.head()
```

```python
# Rename the columns
rename_df.rename(columns={'Mention Title': 'Title', 'Mention URL': 'URL', 'Old_Column3': 'New_Column3'}, inplace=True)

# List of column names to save
columns_to_save = ["Mention Type","Mention Date","Mention Time","Outlet or Author","Title","Country","Subjects","Publication Date","DOI","URL"]

# Create a new DataFrame with only the columns to save
Final = rename_df[columns_to_save]
Final

Final.to_csv('final.csv', index=False)
print("Column names of Rename data set \n", Final.columns.tolist())
Final.info()
Final.head()
```
